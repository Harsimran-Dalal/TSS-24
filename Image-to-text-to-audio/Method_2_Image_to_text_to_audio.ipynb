```python
# Step 2: Import libraries
import pytesseract
from PIL import Image
import gtts
from IPython.display import Audio
```

```python
# Step 3: Open and display the image
img = Image.open('/content/Picture1.png')
print(img)
```

```python
# Step 4: Perform OCR on the image to extract text
result = pytesseract.image_to_string(img)
print(result)

# Step 5: Convert the extracted text to an audio file
tts = gtts.gTTS(result)
tts.save("hello.mp3")

# Step 6: Play the audio file directly in the notebook
Audio("hello.mp3", autoplay=True)
```
